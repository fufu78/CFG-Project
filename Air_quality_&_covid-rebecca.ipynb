```python
# Calculate the average value per location
average_aq = air_quality.groupby('location_code')['av_air_q'].mean().reset_index()
average_aq.rename(columns={'av_air_q': 'average_aq'}, inplace=True)

# Print the new table with average values
print(average_aq)
```

```python
average_aq['average_icu'] = covid_icu_23.iloc[:, 1:8].mean(axis=1)
average_aq['average_deaths'] = covid_deaths_23.iloc[:, 1:8].mean(axis=1)
print(average_aq)
```

# Visualisations of air quality and covid data


## Correlation matric of air quality, icu, deaths

```python
import seaborn as sns

# Create a correlation matrix or any other matrix to visualize
correlation_matrix = average_aq.corr()  # Assuming `average_aq` is a DataFrame

# Create the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')

# Add title and adjust the plot
plt.title('Correlation Heatmap')
plt.tight_layout()

# Display the plot
plt.show()
```

```python
sns.scatterplot(y='average_aq', x='average_icu', hue = 'location_code', data=average_aq, s = 150)
plt.xticks(rotation=90)  # Rotate x-axis labels if needed
plt.title('Average air quality versus average ICU admissions')
plt.ylabel('Average air quality')
plt.xlabel('Average ICU admissions')
plt.show()
```

## Average air quality versus ICU admissions

```python
sns.regplot(x='average_icu', y='average_aq', data=average_aq, color='tomato', scatter_kws={'color': 'dodgerblue'})

correlation_coefficient_aq_test = np.corrcoef(average_aq['average_icu'], average_aq['average_aq'])[0, 1]
correlation_text = f'Correlation (r): {correlation_coefficient_aq_test}'

text_y = average_aq['average_aq'].max() + 20  # Adjust height for placing text

plt.text(average_aq['average_icu'].min(), text_y, correlation_text, color='red', ha='left', va='top')

plt.ylabel('Average air quality (AQI index)')
plt.xlabel('Average ICU/HDU admissions (per 100,000 people)')
plt.title('Average air quality versus average ICU/HDU admissions', pad=20)
plt.show()
```

## Average air quality versus average deaths

```python
sns.scatterplot(y='average_aq', x='average_deaths', hue = 'location_code', data=average_aq, s = 150)
plt.xticks(rotation=90)  # Rotate x-axis labels if needed
plt.ylabel('Average air quality')
plt.xlabel('Average deaths')
plt.title('Average air quality versus average deaths', pad=20)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

sns.regplot(x='average_icu', y='average_aq', data=average_aq, ax=ax1)

correlation_coefficient_aq_test = np.corrcoef(average_aq['average_icu'], average_aq['average_aq'])[0, 1]
correlation_text = f'Correlation (r): {correlation_coefficient_aq_test}'

text_y = average_aq['average_aq'].max() + 20  # Adjust height for placing text

ax1.text(average_aq['average_icu'].min(), text_y, correlation_text, color='red', ha='left', va='top')

ax1.set_ylabel('Average air quality (AQI index)')
ax1.set_xlabel('Average ICU admissions (per 100,000 people)')
ax1.set_title('Average air quality versus average ICU admissions', pad=20)

sns.regplot(x='average_deaths', y='average_aq', data=average_aq, ax=ax2)

correlation_coefficient_aq_test = np.corrcoef(average_aq['average_deaths'], average_aq['average_aq'])[0, 1]
correlation_text = f'Correlation (r): {correlation_coefficient_aq_test}'

text_y = average_aq['average_aq'].max() -5  # Adjust height for placing text

ax2.text(average_aq['average_deaths'].min(), text_y, correlation_text, color='red', ha='left', va='top')

ax2.set_ylabel('Average air quality (AQI index)')
ax2.set_xlabel('Average deaths')
ax2.set_title('Average air quality versus average deaths', pad=20)

plt.tight_layout()
plt.show()
```

## Average ICU admissions and deaths per region

```python
# Convert location_code to numerical representation
average_aq['location_code'] = pd.Categorical(average_aq['location_code'], categories=average_aq['location_code'].unique(), ordered=True)
average_aq['location_code'] = average_aq['location_code'].cat.codes
categories = average_aq['location_code'].unique()

# Set the figure size
plt.figure(figsize=(10, 6)) 

bar_width = 0.35
bar_offset = -bar_width/2

# Create the barplot with adjusted width and hue representing air quality
sns.barplot(x='location_code', y='average_icu', data=average_aq, hue='average_aq', palette='RdYlGn_r', width=bar_width, 
            dodge=False)

plt.legend(title='Air Quality Category')
plt.ylabel('Average ICU/HDU Admissions (per 100,000 people)')
plt.xlabel('Region')

plt.xticks(rotation=90) 
plt.gca().set_xticklabels(regions)

# Create a second y-axis
ax2 = plt.twinx()

# Plot the data on the second y-axis
sns.lineplot(x='location_code', y='average_deaths', data=average_aq, ax=ax2, color='blue')

# Set labels and title


ax2.set_ylabel('Average deaths (total)')
plt.title('Average ICU/HDU admissions (bar) and average deaths (line) per region', pad=20)



plt.show()
```
