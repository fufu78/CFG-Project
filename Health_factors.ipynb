```python
#comparison between both in icu and deaths in 2023

mean_deaths = covid_deaths_23[['deaths_05_feb', 'deaths_12_feb', 'deaths_19_feb', 'deaths_26_feb', 'deaths_05_mar', 'deaths_12_mar', 'deaths_19_mar']].mean()
mean_icu = covid_icu_23[['icu_hdu_rate_per_100000_05_feb', 'icu_hdu_rate_per_100000_12_feb','icu_hdu_rate_per_100000_19_feb','icu_hdu_rate_per_100000_26_feb',
                        'icu_hdu_rate_per_100000_05_mar', 'icu_hdu_rate_per_100000_12_mar','icu_hdu_rate_per_100000_19_mar']].mean()


x = np.arange(len(mean_deaths))


dates = ['05 Feb', '12 Feb', '19 Feb', '26 Feb', '05 Mar', '12 Mar', '19 Mar']

fig, axis1 = plt.subplots(figsize=(10, 6))
axis1.plot(x, mean_deaths, marker='o', linestyle='-', color='blue', label='Mean Deaths')
axis1.set_ylabel('Mean Deaths')


axis2 = axis1.twinx()
axis2.plot(x, (mean_icu*10000), marker='o', color='orange', label='Mean ICU/HDU Rates')
axis2.set_ylabel('Mean ICU/HDU Rates')

axis1.set_xticks(x)
axis1.set_xticklabels(dates)

plt.xlabel('Dates')
plt.title('fig3.1: Comparison of Mean Deaths and Mean ICU/HDU Rates for Each Date')


plt.legend()
plt.show()
```

#### Comparing dates of ICU and deaths

When comparing the deaths and number of people admitted to the ICU, there is no correlation between the two and the dates as the graph lines show different patterns. But in general, there was a low admittance on the 19th of Feb and a high admittance on the 12th of March. 

```python
deaths_05_feb = covid_deaths_23['deaths_05_feb'].tolist()
deaths_12_feb = covid_deaths_23['deaths_12_feb'].tolist()
deaths_19_feb = covid_deaths_23['deaths_19_feb'].tolist()
deaths_26_feb = covid_deaths_23['deaths_26_feb'].tolist()
deaths_05_mar = covid_deaths_23['deaths_05_mar'].tolist()
deaths_12_mar = covid_deaths_23['deaths_12_mar'].tolist()
deaths_19_mar = covid_deaths_23['deaths_19_mar'].tolist()


icu_hdu_rate_per_100000_05_feb = covid_icu_23['icu_hdu_rate_per_100000_05_feb'].tolist()
icu_hdu_rate_per_100000_12_feb = covid_icu_23['icu_hdu_rate_per_100000_12_feb'].tolist()
icu_hdu_rate_per_100000_19_feb = covid_icu_23['icu_hdu_rate_per_100000_19_feb'].tolist()
icu_hdu_rate_per_100000_26_feb = covid_icu_23['icu_hdu_rate_per_100000_26_feb'].tolist()
icu_hdu_rate_per_100000_05_mar = covid_icu_23['icu_hdu_rate_per_100000_05_mar'].tolist()
icu_hdu_rate_per_100000_12_mar = covid_icu_23['icu_hdu_rate_per_100000_12_mar'].tolist()
icu_hdu_rate_per_100000_19_mar = covid_icu_23['icu_hdu_rate_per_100000_19_mar'].tolist()


deaths_data = {
    '05_feb': deaths_05_feb,
    '12_feb': deaths_12_feb,
    '19_feb': deaths_19_feb,
    '26_feb': deaths_26_feb,
    '05_mar': deaths_05_mar,
    '12_mar': deaths_12_mar,
    '19_mar': deaths_19_mar
}

icu_data = {
    '05_feb': icu_hdu_rate_per_100000_05_feb,
    '12_feb': icu_hdu_rate_per_100000_12_feb,
    '19_feb': icu_hdu_rate_per_100000_19_feb,
    '26_feb': icu_hdu_rate_per_100000_26_feb,
    '05_mar': icu_hdu_rate_per_100000_05_mar,
    '12_mar': icu_hdu_rate_per_100000_12_mar,
    '19_mar': icu_hdu_rate_per_100000_19_mar
}

deaths_df = pd.DataFrame(deaths_data)
icu_df = pd.DataFrame(icu_data)


deaths_correlation_matrix = deaths_df.corr().round(2)
icu_correlation_matrix = icu_df.corr().round(2)


plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
sns.heatmap(data=deaths_correlation_matrix, annot=True, cmap='coolwarm')
plt.title('fig3.2: Heatmap for Deaths')

plt.subplot(1, 2, 2)
sns.heatmap(data=icu_correlation_matrix, annot=True, cmap='coolwarm')
plt.title('fig3.3: Heatmap for ICU/HDU Rates')


plt.show()
```

#### Heatmaps for deaths and ICU
The heatmaps show the significance of each date where the red shows high significance and blue shows low. In general there was a higher correlation (red) in february and lower correlation in march for the deaths. 
However, the number of ICU is the opposite where march shows the highest correlation compared to february. 

```python


deaths_data = {
    'location_code': location_codeList,
    'deaths_05_feb': deaths_05_feb,
    'deaths_12_feb': deaths_12_feb,
    'deaths_19_feb': deaths_19_feb,
    'deaths_26_feb': deaths_26_feb,
    'deaths_05_mar': deaths_05_mar,
    'deaths_12_mar': deaths_12_mar,
    'deaths_19_mar': deaths_19_mar
}

icu_data = {
    'location_code': location_codeList,
    'icu_05_feb': icu_hdu_rate_per_100000_05_feb,
    'icu_12_feb': icu_hdu_rate_per_100000_12_feb,
    'icu_19_feb': icu_hdu_rate_per_100000_19_feb,
    'icu_26_feb': icu_hdu_rate_per_100000_26_feb,
    'icu_05_mar': icu_hdu_rate_per_100000_05_mar,
    'icu_12_mar': icu_hdu_rate_per_100000_12_mar,
    'icu_19_mar': icu_hdu_rate_per_100000_19_mar
}

deaths_df = pd.DataFrame(deaths_data)
icu_df = pd.DataFrame(icu_data)


merged_df = deaths_df.merge(icu_df, on='location_code')
correlation_matrix = merged_df.corr().round(2)


plt.figure(figsize=(10, 8))
sns.heatmap(data=correlation_matrix, annot=True, cmap='coolwarm')
plt.title('fig3.4: Correlation Heatmap for Deaths and ICU Rates')
plt.show()
```

#### Heatmaps for deaths vs icu

Based on the heatmap, there is a high correlation between each date in deaths and ICU admittance where majority of the numbers are between 0.56 and 1 showing a red colour.

 Based on the figures generated using data from 2023, we can conclude that some of the dates in 2023 had a higher cases of deaths and ICU admittance. This could be linked to air pollution.
